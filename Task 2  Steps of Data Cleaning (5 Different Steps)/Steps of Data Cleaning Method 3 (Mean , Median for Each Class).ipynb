**After Exploring Data, Now We have To Remove The Columns which have Missing data more than 20%**

```python
missing_values_method_3 = ['Alley', 'FireplaceQu', 'PoolQC', 'Fence', 'MiscFeature']
missing_values_method_3
```

```python
house_dataset_removing_column= house_dataset_method_3.drop(columns=missing_values_method_3)
house_dataset_removing_column.shape
```

**After Removing Missing Values Columns, Now We Select Numerical data to Apply Mean and Median for Each Class**

```python
house_dataset_3_numerical_values = house_dataset_removing_column.select_dtypes(include=['int64', 'float64'])
```

```python
house_dataset_3_numerical_values.shape
```

```python
house_dataset_3_numerical_values.isnull().sum()
```

```python
plt.figure(figsize=(20,5))
sns.heatmap(house_dataset_3_numerical_values.isnull())
```

**As we have Seen Only Three Columns have Missing Values:`LotFrontage', MasVnrArea', GarageYrBlt`. Now We apply Mean value for Each Class. For This Purpose, We will neccessary to Recognize Domain Knowledge**

```python
num_var_miss = ['LotFrontage', 'MasVnrArea', 'GarageYrBlt']
```

```python
house_dataset_method_3.head(2)
```

```python
house_dataset_method_3['LotConfig'].unique()
```

#### For Mean Value

```python
house_dataset_method_3_copy = house_dataset_method_3.copy()
num_vars_miss = ['LotFrontage', 'MasVnrArea', 'GarageYrBlt']
cat_vars = ['LotConfig','MasVnrType','GarageType']
for cat_var, num_var_miss in zip(cat_vars,num_vars_miss):
    for var_class in house_dataset_method_3[cat_var].unique():
        house_dataset_method_3_copy.update(house_dataset_method_3[house_dataset_method_3.loc[:,cat_var] == var_class][num_var_miss].replace(np.nan,house_dataset_method_3[house_dataset_method_3.loc[:,cat_var] == var_class][num_var_miss].mean()))
```

```python
house_dataset_method_3_copy[num_vars_miss].isnull().sum()
```

**Woah! What was that! We apply Mean for Each class but there is Still Missing Values. Let's see what went wrong**

```python
house_dataset_method_3_copy[house_dataset_method_3_copy[['MasVnrArea']].isnull().any(axis=1)]
```

```python
house_dataset_method_3_copy[house_dataset_method_3_copy[['GarageYrBlt']].isnull().any(axis=1)]
```

**Now We have Seen. The Referrence Column is also Empty. That is Why, We have Still Missing Values. Remember! When You getting mean by class, The Refereence column must be Non empty**

```python
house_dataset_method_3_copy = house_dataset_method_3.copy()
num_vars_miss = ['LotFrontage', 'MasVnrArea', 'GarageYrBlt']
cat_vars = ['LotConfig','Exterior2nd','KitchenQual']
for cat_var, num_var_miss in zip(cat_vars,num_vars_miss):
    for var_class in house_dataset_method_3[cat_var].unique():
        house_dataset_method_3_copy.update(house_dataset_method_3[house_dataset_method_3.loc[:,cat_var] == var_class][num_var_miss].replace(np.nan,house_dataset_method_3[house_dataset_method_3.loc[:,cat_var] == var_class][num_var_miss].mean()))
```

```python
house_dataset_method_3_copy[num_vars_miss].isnull().sum()
```

**There We go. After Changing Referrence Column, We have seen that there is No missing Value.**

### Distribution Plot

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(num_vars_miss):
    plt.subplot(2,2,i+1)
    sns.distplot(house_dataset_method_3[var], bins=20, kde_kws={'linewidth':8, 'color':'red'}, label="Original Value",)
    sns.distplot(house_dataset_method_3_copy[var], bins=20, kde_kws={'linewidth':5, 'color':'green'},label="Mean Value",)
    plt.legend()
```

### For Median Value

```python
house_dataset_method_3_copy_median = house_dataset_method_3.copy()
num_vars_miss = ['LotFrontage', 'MasVnrArea', 'GarageYrBlt']
cat_vars = ['LotConfig','Exterior2nd','KitchenQual']
for cat_var, num_var_miss in zip(cat_vars,num_vars_miss):
    for var_class in house_dataset_method_3[cat_var].unique():
        house_dataset_method_3_copy_median.update(house_dataset_method_3[house_dataset_method_3.loc[:,cat_var] == var_class][num_var_miss].replace(np.nan,house_dataset_method_3[house_dataset_method_3.loc[:,cat_var] == var_class][num_var_miss].median()))
```

```python
house_dataset_method_3_copy_median[num_vars_miss].isnull().sum()
```

### Plot Distribution For Both Mean and Median Values

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(num_vars_miss):
    plt.subplot(2,2,i+1)
    sns.distplot(house_dataset_method_3[var], bins=20, kde_kws={'linewidth':8, 'color':'red'}, label="Original Value",)
    sns.distplot(house_dataset_method_3_copy[var], bins=20, kde_kws={'linewidth':5, 'color':'green'},label="Mean Value",)
    sns.distplot(house_dataset_method_3_copy_median[var], bins=20, kde_kws={'linewidth':5, 'color':'Blue'},label="Median Value",)
    plt.legend()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(num_vars_miss):
    plt.subplot(2,2,i+1)
    plt.hist(house_dataset_method_3[var],label="Original Value",color='Blue')
    plt.hist(house_dataset_method_3_copy[var],label="Mean Value",color='Green')
    plt.hist(house_dataset_method_3_copy_median[var],label="Median Value",color='Red')
    plt.legend()
```

### For Checking Outliers In Dataset

```python
for i, var in enumerate(num_vars_miss):
    plt.figure(figsize=(10,10))
    plt.subplot(3,1,1)
    sns.boxplot(house_dataset_method_3[var])
    plt.subplot(3,1,2)
    sns.boxplot(house_dataset_method_3_copy[var])
    plt.subplot(3,1,3)
    sns.boxplot(house_dataset_method_3_copy_median[var])
```
