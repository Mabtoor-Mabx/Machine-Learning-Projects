## Steps of Data Cleaning Method 1 (Deleting Rows and Columns)

**Data cleansing or data cleaning is the process of detecting and correcting corrupt or inaccurate records from a record set, table, or database and refers to identifying incomplete, incorrect, inaccurate or irrelevant parts of the data and then replacing, modifying, or deleting the dirty or coarse data.**

**We Have to Clean Dataset Using Method Deleting Rows and Columns**

```python
# Import Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Dataset 

house_dataset_method_1 = pd.read_csv(r'train.csv')
house_dataset_method_1.head()
```

```python
house_dataset_method_1.tail()
```

```python
house_dataset_method_1.head(5)
```

### Explore data for Cleaning

```python
house_dataset_method_1.info()
```

```python
house_dataset_method_1.describe()
```

```python
house_dataset_method_1.isnull().sum()
```

```python
# Plot a Graph

plt.figure(figsize=(20,5))
sns.heatmap(house_dataset_method_1.isnull())
```

```python
# To Check Numerical Values

num_values = house_dataset_method_1.isnull().sum()/house_dataset_method_1.shape[0] * 100
num_values
```

**As we have Seen There is Some Missing data. So , We Drop the Columns Which Have missing data greater than 17%**

```python
drop_columns = num_values[num_values > 17].keys()
drop_columns
```

```python
house_dataset_method_1_removing_column = house_dataset_method_1.drop(columns=drop_columns)
```

```python
house_dataset_method_1_removing_column.isnull().sum()
```

```python
plt.figure(figsize=(20,5))
sns.heatmap(house_dataset_method_1_removing_column.isnull())
```

**As We have Seen After Removing column, The Data is Better than previous. Now We have to Remove Rows so that our data will be clean**

```python
house_dataset_method_1_clean_data = house_dataset_method_1_removing_column.dropna()
```

```python
house_dataset_method_1_clean_data.isnull().sum()
```

```python
plt.figure(figsize=(20,5))
sns.heatmap(house_dataset_method_1_clean_data.isnull())
```

**Now We have Seen That our Data is Clean. Its Time To plot the Distribution to compare original and Clean data**

###### Create Distribution Plot for Numerical Data

```python
house_dataset_method_1_clean_data.select_dtypes(include=['int64', 'float64']).columns
```

```python
sns.distplot(house_dataset_method_1['MSSubClass'])
sns.distplot(house_dataset_method_1_clean_data['MSSubClass'])
```

```python
numerical_variables= ['MSSubClass', 'LotArea', 'OverallQual', 'OverallCond',
       'YearBuilt', 'YearRemodAdd', 'MasVnrArea', 'BsmtFinSF1', 'BsmtFinSF2',
       'BsmtUnfSF', 'TotalBsmtSF', '1stFlrSF', '2ndFlrSF', 'LowQualFinSF',
       'GrLivArea', 'BsmtFullBath', 'BsmtHalfBath', 'FullBath', 'HalfBath',
       'BedroomAbvGr', 'KitchenAbvGr', 'TotRmsAbvGrd', 'Fireplaces',
       'GarageYrBlt', 'GarageCars', 'GarageArea', 'WoodDeckSF', 'OpenPorchSF',
       'EnclosedPorch', '3SsnPorch', 'ScreenPorch', 'PoolArea', 'MiscVal',
       'MoSold', 'YrSold', 'SalePrice']

plt.figure(figsize=(25,25))       
for i,var in enumerate(numerical_variables):
    plt.subplot(9,4,i+1)
    sns.distplot(house_dataset_method_1[var], bins=20)
    sns.distplot(house_dataset_method_1_clean_data[var], bins=20)
```

```python
plt.figure(figsize=(25,25))
for i,var in enumerate(numerical_variables):
    plt.subplot(9,4,i+1)
    plt.hist(house_dataset_method_1[var], label='Original Data')
    plt.hist(house_dataset_method_1_clean_data[var], label='Clean Data')
    plt.legend()
```

###### Comparison of Categorical Data

```python
house_dataset_method_1_clean_data.select_dtypes(include=['object']).columns
```

```python
pd.concat([house_dataset_method_1['MSZoning'].value_counts()/house_dataset_method_1.shape[0] * 100, 
          house_dataset_method_1_clean_data['MSZoning'].value_counts()/house_dataset_method_1_clean_data.shape[0] * 100], axis=1,
         keys=['MSZoning_org', 'MSZoning_clean'])
```

```python
def categorical_values_distribution(var):
    return pd.concat([house_dataset_method_1[var].value_counts()/house_dataset_method_1.shape[0] * 100, 
          house_dataset_method_1_clean_data[var].value_counts()/house_dataset_method_1_clean_data.shape[0] * 100], axis=1,
          keys=[var+'_org', var+'clean'])
```

```python
print(categorical_values_distribution('MSZoning'),'\n')
print(categorical_values_distribution('Street') ,'\n')
print(categorical_values_distribution('LotShape') ,'\n')
print(categorical_values_distribution('LandContour') ,'\n')
print(categorical_values_distribution('Utilities') ,'\n')
print(categorical_values_distribution('BsmtCond') ,'\n')
print(categorical_values_distribution('ExterQual'),'\n')
print(categorical_values_distribution('BsmtFinType1') ,'\n')
print(categorical_values_distribution('GarageType') ,'\n')
print(categorical_values_distribution('SaleCondition') ,'\n')
```
