# Steps of Data Cleaning Method 2 (Missing Value Imputation by Mean and Median)
**Data cleansing or data cleaning is the process of detecting and correcting corrupt or inaccurate records from a record set, table, or database and refers to identifying incomplete, incorrect, inaccurate or irrelevant parts of the data and then replacing, modifying, or deleting the dirty or coarse data.**

**We Have to Clean Dataset Using Method Missing Value Imputation by Mean and Median**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Cleaning The Data

```python
missing_value_percentage = house_dataset_method_2.isnull().sum() / house_dataset_method_2.shape[0] * 100
missing_value_percentage
```

**We have to remove The columns which have missing values greater than 17%**

```python
missing_values_columns = missing_value_percentage[missing_value_percentage > 17].keys()
```

```python
house_dataset_removing_columns = house_dataset_method_2.drop(columns = missing_values_columns)
```

```python
house_dataset_removing_columns.isnull().sum()
```

```python
plt.figure(figsize=(20,5))
sns.heatmap(house_dataset_removing_columns.isnull())
```

**As we have seen after removing columns which have missing values greater than 17%, Our dataset is getting better. Now we have to get mean and median value and apply it in remaining missing value**

**But Before mean and Median, We have to select The Numerical Values**

```python
house_dataset_numerical_values = house_dataset_removing_columns.select_dtypes(include=['int64', 'float64'])
```

```python
house_dataset_numerical_values.head()
```

```python
house_dataset_numerical_values.isnull().sum()
```

```python
plt.figure(figsize=(20,5))
sns.heatmap(house_dataset_numerical_values.isnull())
```

```python
missing_numerical_variable = [var for var in house_dataset_numerical_values.columns if house_dataset_numerical_values[var].isnull().sum()>0]
missing_numerical_variable
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(missing_numerical_variable):
    plt.subplot(2,2,i+1)
    sns.distplot(house_dataset_numerical_values[var], bins=20, kde_kws={'linewidth':5, 'color':'#DC143C'})
```

```python
house_dataset_mean = house_dataset_numerical_values.fillna(house_dataset_numerical_values.mean())
house_dataset_mean.head(2)
```

```python
house_dataset_mean.isnull().sum().sum()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(missing_numerical_variable):
    plt.subplot(2,2,i+1)
    sns.distplot(house_dataset_numerical_values[var], bins=20, kde_kws={'linewidth':5, 'color':'blue'})
    sns.distplot(house_dataset_mean[var], bins=20, kde_kws={'linewidth':5, 'color':'green'})
```

```python
house_dataset_median = house_dataset_numerical_values.fillna(house_dataset_numerical_values.median())
```

```python
house_dataset_median.head(2)
```

```python
house_dataset_median.isnull().sum().sum()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(missing_numerical_variable):
    plt.subplot(2,2,i+1)
    sns.distplot(house_dataset_numerical_values[var], bins=20, kde_kws={'linewidth':5, 'color':'blue'})
    sns.distplot(house_dataset_mean[var], bins=20, kde_kws={'linewidth':5, 'color':'green'})
    sns.distplot(house_dataset_median[var], bins=20, kde_kws={'linewidth':5, 'color':'red'})
```

### For Checking Outliers

```python
for i, var in enumerate(missing_numerical_variable):
    plt.figure(figsize=(10,10))
    plt.subplot(3,1,1)
    sns.boxplot(house_dataset_method_2[var])
    plt.subplot(3,1,2)
    sns.boxplot(house_dataset_mean[var])
    plt.subplot(3,1,3)
    sns.boxplot(house_dataset_median[var])
```

```python
house_dataset_concat = pd.concat([house_dataset_numerical_values[missing_numerical_variable], house_dataset_mean[missing_numerical_variable], house_dataset_median[missing_numerical_variable]], axis=1)
```
